#  Beyond cumulative risk and dimensional models: A data driven approach to examining the dimensional- interactive effects of Early Life Stress on brain structure and externalizing psychopatholpogy

## Background

Over half of all youth will be exposed to at least one form of early life stress, such as abuse, violence exposure, neglect, divorce, or poverty, before reaching adulthood (Green et al., 2010; McLaughlin et al., 2012). Early life stress (ELS) has been robustly associated with increased risk for a host of behavioral difficulties across development, including aggression, oppositional behaviors, impulsivity, and other externalizing problems (Benjet et al., 2010; Doom et al., 2016, 2016; Font & Berger, 2015; Harford et al., 2014; Herts et al., 2012; McLaughlin et al., 2012; Moylan et al., 2010; Piotrowska et al., 2019). Despite this significant body of work, the mechanisms underlying these associations remain unclear. Recently, research has turned to using multiple levels of analysis (i.e., brain structure and function) to better elucidate the links between ELS and increased risk for externalizing psychopathology. However, progress has been modest, potentially due to dominating conceptual models of ELS, that is, cumulative and multiple risk models, that offer high predictability at the expense of specificity. Initial progress has been made to increase specificity in the modeling of ELS with dimensional models of adversity exposure (i.e., Deprivation/Threat; McLaughlin & Sheridan, 2016; and Harshness/Unpredictability; Belsky et al., 2012). However, while the deprivation-threat model has informed a considerable body of work examining associations between ELS and neurodevelopmental outcomes, there is a pressing need for studies examining the impact of unpredictability in childhood rearing environments on neural development, as this area remains virtually unexplored in the current literature (McLaughlin et al., 2019). The present proposal aims to address these gaps by integrating both multiple risk and dimensional models of adversity exposure through a data dirven approach to modelng ELS in association with hippocampal volume in a community sample fo youth aged 9 -18 years. 

The multiple risk model introduced by Rutter and colleagues in the late seventies rests upon the assumpton that that above and beyond any single risk factor alone, the joint occurrence of multiple risk factors (e.g., parental psychopathology, crowding, parental occupational status) potentiate risk for the development of problem behaviors (Rutter, 1979; Rutter, Cox, et al., 1975; Rutter, Yule, et al., 1975). While Rutter’s original work demonstrated a multiplicative (i.e., interactive) effect of additional exposures on the risk for problem behaviors, subsequent applications of the model additionally revealed evidence of linear relationship, with greater exposure to familial and socioeconomic stressors during infancy, childhood, and adolescence predicting increased prospective risk for the development of externalizing behaviors (Blanz et al., 1991; Shaw et al., 1994; Shaw & Emery, 1988; Trentacosta et al., 2008). These findings along with increasing evidence documenting the high rate of co-occurrence among ELS exposures (i.e., a child who endorses one exposure is likely to endorse having experienced multiple exposures) resulted in the proliferation of research using cumulative risk models of ELS exposure, which focus on the additive sum of exposures rather than their individual or interactive effects (Arata et al., 2007; Dong, Anda, et al., 2004; Dube et al., 2003; Edwards et al., 2003; Finkelhor et al., 2007).  
   
   
In the cumulative risk approach, individual stress exposures (e.g., poverty, crowding, noise pollution, parental psychopathology, community violence, sexual abuse) are dichotomized (i.e., 0=not exposed, 1=exposed) and subsequently summed to yield a cumulative risk score (Evans et al., 2013). The cumulative risk approach, much like its predecessor the multiple risk model, has been noted primarily for its strong statistical explanatory power, as the potent dose-response relationship between number of ELS exposures and increased risk for psychopathology is well documented, with higher cumulative risk scores being linked to higher aggressive, and oppositional behaviors as well as higher rates of oppositional defiant disorder (ODD), conduct disorder (CD), ADHD, and substance use disorders across adolescence and young adulthood (Aebi et al., 2014; Appleyard et al., 2005; Doan et al., 2012; Edwards et al., 2003; Gerard & Buehler, 2004; Kessler et al., 2010; McLaughlin et al., 2012). 

A major limitation of the cumulative and multiple risk approaches, however, is that individual exposures are often summed regardless of type, severity, or timing of exposure, which therefore presumes that all exposures impact development in equivalent manners (Belsky et al., 2012; McLaughlin & Sheridan, 2016; Nelson & Gabard-Durnam, 2020; Zeanah & Sonuga‐Barke, 2016). This lack of specificity has prevented its application to developmental neuroscience based research in which both the saliency and frequency of environmental input as well the timing as it relates to plasticity, our fundamental constructs for understanding neural developmental, particularly with regard to understanding neurodedvelopmental trajectories for cognitive development.
   
Dimensional models of adversity exposure emerged to inform neuroscience research focused on identifying mechanistic pathways leading to increased risk. One approach, the deprivation-threat model, introduced by McLaughlin and Sheridan (2016), proposes deprivation and threat as two orthogonal dimensions of environmental experience. Accordingly, each dimension is hypothesized to underlie multiple forms of adversity exposure that are likely to impact neural development is similar ways. Deprivation is characterized as the absence of expected inputs (e.g., physical neglect, psychosocial deprivation, and cognitive deprivation), whereas threat is characterized by experiences of harm or expected harm (physical abuse, domestic violence, and community violence; McLaughlin et al., 2014; McLaughlin & Sheridan, 2016).
   
Importantly, experiences of threat and deprivation are theorized to have at least partially distinct impacts on neural development (McLaughlin & Sheridan, 2016). Threat is posited to primarily impact neural systems involved in fear acquisition, emotional salience processing, and emotion regulation, namely fronto-limbic circuits including the medial prefrontal cortex (mPFC), amygdala, and hippocampus, while deprivation is posited to primarily impact frontoparietal systems underlying cognitive development (McLaughlin et al., 2019). While there has been some support for the deprivation-threat model, with most studies demonstrating heightened amygdala activity in children exposed to threat, results among studies examining deprivation are mixed, with inconsistent findings regarding frontoparietal activity and connectivity (McLaughlin et al., 2019). 
 
    
While the deprivation-threat model has facilitated considerable progress, it is not comprehensive. An alternative approach put forth by Belsky and colleagues (2012) is the harshness/unpredictability model, based on life history theories of development. Belsky and colleagues posit that evolutionarily driven responses for adaptation and survival drive the developmental sequalae of ELS. Harshness refers to environmental experiences that signal morbidity or mortality (e.g., harsh parenting or neighborhood conditions), whereas unpredictable environmental experiences signal resource scarcity (e.g., residential changes or parental job transitions; Belsky et al., 2012). While harsh experiences indicate threat to immediate survival, unpredictable experiences indicate uncertainty in the availability of future resources needed for survival. Thus, both types of experience bias development towards prioritizing behaviors that capitalize on immediate resources and benefits over those relating to more long-term outcomes or investments, which may promote greater impulsivity and risk-taking behaviors (i.e., “fast” life history strategies; Belsky et al., 2012; Ross & Hill, 2002).
    
    
Evidence for the distinct influences of harshness and unpredictability on development has been inconsistent and thus inconclusive (Jonason et al., 2016; Simpson et al., 2012; Warren & Barnett, 2020). However, a notable number of findings do demonstrate associations between unpredictability (i.e., paternal, residential, and parental job transitions) and higher rates of risk-taking behaviors and externalizing problems across development (Belsky et al., 2012; Chang et al., 2019; Doom et al., 2016; Hartman et al., 2018; Simpson et al., 2012). While associations between unpredictability and neural development in humans remain unexplored, emerging work suggests that unpredictable maternal sensory signals may be related to disruptions in hippocampal-dependent memory formation and effortful control, suggestive of potential neurodevelopmental pathways implicated in increased impulsivity, risk-taking, and externalizing behaviors (Davis et al., 2017, 2019), that is specifc to the hippocampus.

    

Based on prior empircal work, the present study integrates dimensional models of ELS with risk based models by conducting a dimensionally informed model selection in an model that examines the interaction between exposures and their direct and indirect effects on increased risk for externalizing psychopathology.

The current study will conduct a moderated mediation analyssis in which models will be examined with respect to subset of interacting terms informed by dimensional models.

### Model: Indirect Effects

$$ \hat{M_{hv}} = i_1 + \hat{a_1}X_{id} + \hat{a_2} U_{sex} + \hat{a_3} U_{age} + \hat{a_4} U_{wbv}$$


$$ \hat{Y_{ext}} = i_2 + \hat{{c'}_1}X_{id} + \hat{b}{M_{hv}} + \hat{{c'}_2} U_{sex} + \hat{{c'}_3} U_{age} + \hat{{c'}_4} U_{wbv}$$

Where, ${a_1}b$ = indirect effect of $X_{id}$ on$Y_{ext}$ operating through {M_{hv}}

and ${{c'}_3}$ = conditional direct effect of $X_{id}$ on $Y_{ext}$, independent of {M_{hv}}


### Moderated Mediation Model 

$$ \hat{M_{hv}} = i_1 + \hat{a_1}X_{id} + \hat{a_2}W_{mod} + \hat{a_3}X_{id}W_{mod} + \hat{a_4}U_{sex} + \hat{a_5} U_{age} + \hat{a_6} U_{wbv}$$


$$ \hat{Y_{ext}} = i_2 + \hat{{c'}_1}X_{id} + \hat{{c'}_2}W_{mod} + hat{{c'}_3}X_{id}W_{mod}+ \hat{b}{M_{hv}} + \hat{{c'}_4} U_{sex} + \hat{{c'}_5} U_{age} + \hat{{c'}_6} U_{wbv}$$



Where, ${a_3}b$ = index of moderated mediation 

and ${{c'}_3}$ = conditional direct effect of  $X_{id}$ on$Y_{ext}$

### Tidying and Harmonization of Hippocampal Volume Data 

```r
#Loading libraries
library(easypackages)
libraries("tidyverse","skimr", "naniar", "stringi", "psych", "DataExplorer", "boot", "visdat", "VIM" )
#Loading hbn phenotypic data to be included in harmonization (aka-- age &  sex); 
#files were taken from http://fcon_1000.projects.nitrc.org/indi/cmi_healthy_brain_network/sharing_neuro.html
pheno.d = read_csv('HBN_R1_1_Pheno.csv')%>%
  rbind(read_csv('HBN_R2_1_Pheno.csv'))%>%
  rbind(read_csv('HBN_R3_Pheno.csv'))%>%
  rbind(read_csv('HBN_R4_Pheno.csv'))%>%
  rbind(read_csv('HBN_R5_Pheno.csv'))%>%
  rbind(read_csv('HBN_R6_Pheno.csv'))%>%
  rbind(read_csv('HBN_R7_Pheno.csv'))%>%
  rbind(read_csv('HBN_R8_Pheno.csv'))%>%
  rename(sub_id = EID)%>%
  select_all(tolower)%>%
  #removing duplicate subjects 
  .[!duplicated(.$sub_id), ]
```

```r
#loading hippocampal volume data
hippo.d = read_csv("hbn_hippo_vols.csv")%>%
  #tidying data!
  separate(site_wave, into = c("site", "wave"), sep = "_")%>%
  mutate(sub_id = str_remove(sub_id, "sub-"))%>%
  #merging with phenotypic data
  left_join(pheno.d, by = "sub_id")%>%
  #removing uneccessary variables 
  select(-wave, -ehq_total, -commercial_use, -full_pheno)
#Setting up transposed matrices [p rows (features) x n columns (subjects)] for harmonization of structural data across sites
#Subject id vector 
sub_id = hippo.d[,1]
# Batch <- site = {CBIC, SI, RU}
batch = t(hippo.d[,2])
# dat <- whole_brain_vol, hippo_vol_lh, hippo_vol_rh
dat = t(hippo.d[,3:5])
#mod <- age, sex 
  #mod is matrix with [n rows (subjects) x p columns (features)] 
mod = as.matrix(hippo.d[,c("age","sex")])
#Running combat harmonization on hbn hippocampal volume data 
#Sourcing combat scripts
source("Combat/R/scripts/utils.R")
source("Combat/R/scripts/combat.R")
#Harmonizing data
hippo.harmonized <- combat(dat=dat, batch=batch, mod=mod)
#Transposing harmonized data to get a [nxp] matrix
hippo.d = t(hippo.harmonized$dat.combat)
#Setting up data.frame with sub_id and harmonized data
hippo.d = data.frame(sub_id = sub_id, 
                     whole_brain_vol = hippo.d[,1], 
                     hippo_vol_lh = hippo.d[,2], 
                     hippo_vol_rh = hippo.d[,3])
hippo.d = hippo.d%>% mutate(hippo_vol = hippo_vol_rh + hippo_vol_lh)
```

### Phenotypic Data: Tidying

1. Tidying dataset 
2. Filtering based on exclusion criteria 
3. Merging with Hippocampal data 

```r
#CLEANSING PHENOTYPIC DATA
#creating vector of suffixes of nonsense variables that will be removed during cleansing
nonsense = c(",Administration", ",Comment_ID", ",Data_entry", ",Days_Baseline",
             ",EID",",PSCID",",Year", ",Patient_ID",",START_DATE",",Season", 
             ",Site",",Study", "Visit_label",",Days_Baseline")
#Sourcing file with my functions to be used during cleansing and processing 
  # ~p1p2 = used to take the higher score or value between two reporters (e.g., taking the higher of the two scores between parent and child report)
  # ~myrecode = a simple recoding function. no bells and whistles. easy to use. primarily used for reverse coding of items.(e.g., recoding values from c(1,2,3) to c(3,2,1))
source("dspan_final_project_functions.R")

#Loading HBN phenotypic data file
  #Setting periods & blanks as NaNs
hbn.d = read_csv("hbn_dspan_phenotypic_data.csv", na = c("", "."))%>%
  #rename column name "identifiers" to "sub_id" 
  rename(sub_id = Identifiers
  )%>% 
  #remove 2nd row (does not contain any data)
  subset(sub_id != ",assessment"
  )%>%
  #remove ",assessment" from the end of each sub_id character string so can merger later with hippocampal volume data
  mutate(sub_id = str_remove(sub_id, ",assessment")
  )%>%
  #remove all nonsense variables from dataset (they are repeated for each measure and largely do not contain any data,
  #or if they do, none that will be used in the current analysis)
  select(-matches(paste0('.*', nonsense, '$', collapse = '|'), ignore.case = FALSE)
  )%>%
  #Editing column names to remove commas and redundant prefix (e.g., chnaging APQ,APQ_ID to APQ_ID)
    #And setting all column names to lowercase (personal preference) 
  select_all(~gsub(".*,", "", tolower(.)))%>%
#Covariates and exclusion criteria
  #IQ: collapsing across various IQ test scores to create single IQ variable
  #Sex: converting sex to a factor 
  mutate(waiswasi_fsqi = ifelse(is.na(wasi_fsiq), wais_fsiq_scale, wasi_fsiq),
         iq = ifelse(!is.na(wisc_fsiq), wisc_fsiq,waiswasi_fsqi),
         sex = as.factor(sex)
         )%>%
  #Filtering based on age range of interest (late childhood through adolescence) and exclusion criteria of IQ<70
  filter(age >= 9, age<19, iq>=70)%>% 
  #Merging with harmonized hippocampal data; only keeping subjects with hippocampal volume data
  right_join(hippo.d, by= "sub_id")%>%
  #Filtering to remove subjects missing data for the other primary variables in mediation model (X,M)
  filter(!is.na(apq_p_id), !is.na(cbcl_ext_t))%>% 
#Doing a little renaming before auditing so the names actually make sense
  #Threat
  rename(nles_sr_dv = nles_sr_06a)%>%
  #SES 
  rename(fsq_income = fsq_04, fsq_wic = fsq_05h, 
         fsq_snap = fsq_05i,  fsq_nyccash = fsq_05j, 
         housetenure = fsq_06)%>%
  rename(nle_sr_fstrain = nles_sr_09a, nle_sr_jobloss = nles_sr_17a,
         nle_p_fstrain = nles_p_08a, nle_p_jobloss = nles_p_15a)%>%
 #Parental Absence 
  rename(nle_sr_jail = nles_sr_16a, nle_p_jail= nles_p_14a, p2_absence = no_p2)%>%
 #Parring the dataset down what is needed
  select(sub_id, age, sex, whole_brain_vol, apq_p_id, hippo_vol, cbcl_ext_t,
         cpic_01:cpic_51, apq_p_cp, nles_sr_dv, fsq_income, fsq_wic, fsq_snap, 
         fsq_nyccash, barratt_p1_edu, barratt_p2_edu, 
         nle_sr_fstrain, nle_p_fstrain, nle_sr_jobloss, nle_p_jobloss, housetenure, 
         apq_p_pm, nle_sr_jail, nle_p_jail, p2_absence, psi_pd_t,
         psi_total_t, fattempts, mattempts, 
         mdx, m_dep, m_anxiety, m_panic, m_ocd, m_alc_abuse, m_drug_abuse, m_bipolar, 
         m_schiz, mdxpsych, fdx, f_dep, f_anxiety, f_panic, f_ocd, f_alc_abuse, 
         f_drug_abuse, f_bipolar, f_schiz, fdxpsych)
head(hbn.d)
```

## Recoding remaining variables. Refer to tables 1 and 2 for details.

```r

#Recoding Variables
#Threat 
  #Recoding CPIC so that higher scores reflect worse outcomes
    #using myrecode function
 hbn.d <- hbn.d %>% mutate_at(vars(cpic_01:cpic_51), ~myrecode(., old = c(0,1,2), new = c(2,1,0)))%>%
  mutate(cpic_intensity = cpic_05 + cpic_14 + cpic_24 + cpic_33 + cpic_38 + cpic_40 +cpic_45,
         cpic_freq = cpic_01 + cpic_10 + cpic_16 + cpic_20 + cpic_29 + cpic_37,
         cpic_perc_threat = cpic_07 + cpic_17 + cpic_26 + cpic_35 + cpic_42 + cpic_47)%>%

#SES variables
  #Parent Education: 
    #using p1p2 function to set parent education equal to higher reported value between parent 1 and parent 2
    #If one of the two values is missing, then parent education is set to equal whichever value is present 
    #(e.g., if p1 = 7, p2 = NA, then pedu = 7)
  mutate(pedu = p1p2(barratt_p1_edu, barratt_p2_edu, scoreas = "higher"))%>%
  #Financial hardship: 
    #collapsing across 2 Negative Life Events Scale items to create dummy variable indicating endorsement of financial hardship
    #1 = responded yes to both the financial strain and job loss items
    #0 = response of "no" to either the financial strain and job loss item 
      #financial strain: mom or dad talked about having serious money troubles (being worried about bills for ordinary things)
      #job loss: mom or dad lost their job
  mutate(fnhardship_p = case_when((is.na(nle_p_fstrain) & is.na(nle_p_jobloss)) ~ nle_p_fstrain,
                                           (nle_p_fstrain==1 & nle_p_jobloss==1) ~ 1,
                                           TRUE ~ 0))%>%
  #Income:
    #recoding "choose not to disclose" as missing
    #changing range for income scores from 0-11 to 1-12 (preferable scale as values of 0 are typically used to indicate "nothing", "no", or "none") 
    #creating continuous measure of income based on midpoint of income bin (refer to Appendix 1 for description of variables)
    #taking natural log of income to provide some adjustment for severe right skew 
  replace_with_na(replace = list(fsq_income = 12))%>%
  mutate(income = fsq_income +1,
         income_cont = 
           case_when(is.na(income) ~ income, 
                     income == 1   ~ 5000, 
                     income == 2   ~ 15000,
                     income == 3   ~ 25000,
                     income == 4   ~ 35000,
                     income == 5   ~ 45000, 
                     income == 6   ~ 55000, 
                     income == 7   ~ 65000, 
                     income == 8   ~ 75000, 
                     income == 9   ~ 85000,
                     income == 10  ~ 95000,
                     income == 11  ~ 125000,
                     income == 12  ~ 150000),
         ln_income_cont = log(income_cont)
         )%>%
  #Welfare status: 
    #dichotomously coded variable (0=no/1=yes) indicating endorsement of eligibility for or receipt of 
    #welfare/income based federal financial support for/from any the following programs:
        #Special Supplemental Nutrition Program for Women, Infants, and Children (WIC),
        #Supplemental Nutrition Assistance Program (SNAP) Benefits
        #NYC Cash Assistance Program
  #NOTE coding done inclusively such that only NA on all variables yielded NA for collapsed variable
mutate(fed_income = case_when((is.na(fsq_wic) & is.na(fsq_snap) & is.na(fsq_nyccash)) ~ fsq_wic,
                   (fsq_wic==1 | fsq_snap==1 | fsq_nyccash==1)             ~ 1,
                   (fsq_wic==0 | fsq_snap==0 | fsq_nyccash==0)             ~ 0))%>%
  #Housing tenure status: 
    #recoding "2 = choose not to disclose" as missing
    #0 = own, 1 = rent 
  replace_with_na(replace = list(housetenure = 2))%>%
  #Parental Absence
  #Selecting variables for analysis 
  select(sub_id, age, sex, whole_brain_vol, hippo_vol, cbcl_ext_t, apq_p_id,  fnhardship_p,
         pedu, ln_income_cont, income_cont, fed_income, housetenure,
        apq_p_pm, nle_p_jail, 
         apq_p_cp, 
         cpic_freq, cpic_intensity, cpic_perc_threat, psi_total_t,
        psi_pd_t, pattempts_any, pattempts_thresh,
         m_mood_disorder, msubstance, mpsychosis, m_psychopathology, 
         f_mood_disorder, fsubstance, fpsychosis, f_psychopathology, 
         p_mood_disorder, psubstance, ppsychosis, p_psychopathology) 
  
```

```r

#MODERATED MEDIATION 
momed.fn = function (data, index) {
  effs <- NULL
  aics <- NULL
  X <- data$apq_p_id
  M <- data$hippo_vol
  Y <- data$cbcl_ext_t
  U1 <- data$age
  U2 <- data$sex
  U3 <- data$whole_brain_vol
  #functions to be called
  aic.fn <- function(model){aic <-length(model$residuals)*(log(sum(model$residuals^2)/length(model$residuals)))+ (2*length(model$coefficients -1))}
  momed.eff.fn <- function(a,b){
    m.index <- a$coefficients[7]*b$coefficients[4]
    cond.direct <- b$coefficients[8]
             effs <- c(m.index = m.index , cond.direct = cond.direct)
  }
  
  #Models
  #Original Mediation Model
  a.med = lm(M ~ X + U1 + U2 + U3, data=data, subset=index, na.action = na.omit)
  b.med = lm(Y ~ X  + M + U1 + U2 + U3, data=data, subset=index, na.action = na.omit)
  med.model <- c(a.med = a.med$coefficients, 
                 b.med=b.med$coefficients, 
                 indirect.med = a.med$coefficients[2]*b.med$coefficients[3], 
                 direct.med = b.med$coefficients[2], 
                 aic.med = aic.fn(b.med))
  
  a.fnhardship_p =  lm(M ~ X + fnhardship_p*X + U1 + U2 + U3, data=data, subset=index, na.action = na.omit)
  b.fnhardship_p = lm(Y ~ X  + fnhardship_p*X + M + U1 + U2 + U3,  data=data, subset=index, na.action = na.omit)
  eff <- momed.eff.fn(a.fnhardship_p,b.fnhardship_p )
  effs <- c(effs, eff)
  aic <- c(fnhardship_p.aic = aic.fn(b.fnhardship_p))
  aics <- c(aics, aic)
  a.pedu =  lm(M ~ X + pedu*X + U1 + U2 + U3, data=data, subset=index, na.action = na.omit)
  b.pedu = lm(Y ~ X  + pedu*X + M + U1 + U2 + U3,  data=data, subset=index, na.action = na.omit)
  eff <- momed.eff.fn(a.pedu,b.pedu )
  effs <- c(effs, eff)
  aic <- c(pedu.aic = aic.fn(b.pedu))
  aics <- c(aics, aic)
  a.ln_income_cont =  lm(M ~ X + ln_income_cont*X + U1 + U2 + U3, data=data, subset=index, na.action = na.omit)
  b.ln_income_cont = lm(Y ~ X  + ln_income_cont*X + M + U1 + U2 + U3,  data=data, subset=index, na.action = na.omit)
  eff <- momed.eff.fn(a.ln_income_cont,b.ln_income_cont )
  effs <- c(effs, eff)
  aic <- c(ln_income_cont.aic = aic.fn(b.ln_income_cont))
  aics <- c(aics, aic)
  a.fed_income =  lm(M ~ X + fed_income*X + U1 + U2 + U3, data=data, subset=index, na.action = na.omit)
  b.fed_income = lm(Y ~ X  + fed_income*X + M + U1 + U2 + U3,  data=data, subset=index, na.action = na.omit)
  eff <- momed.eff.fn(a.fed_income,b.fed_income )
  effs <- c(effs, eff)
  aic <- c(fed_income.aic = aic.fn(b.fed_income))
  aics <- c(aics, aic)
  a.housetenure =  lm(M ~ X + housetenure*X + U1 + U2 + U3, data=data, subset=index, na.action = na.omit)
  b.housetenure = lm(Y ~ X  + housetenure*X + M + U1 + U2 + U3,  data=data, subset=index, na.action = na.omit)
  eff <- momed.eff.fn(a.housetenure,b.housetenure )
  effs <- c(effs, eff)
  aic <- c(housetenure.aic = aic.fn(b.housetenure))
  aics <- c(aics, aic)
  
  a.apq_p_pm =  lm(M ~ X + apq_p_pm*X + U1 + U2 + U3, data=data, subset=index, na.action = na.omit)
  b.apq_p_pm = lm(Y ~ X  + apq_p_pm*X + M + U1 + U2 + U3,  data=data, subset=index, na.action = na.omit)
  eff <- momed.eff.fn(a.apq_p_pm,b.apq_p_pm )
  effs <- c(effs, eff)
  aic <- c(apq_p_pm.aic = aic.fn(b.apq_p_pm))
  aics <- c(aics, aic)
  a.nle_p_jail =  lm(M ~ X + nle_p_jail*X + U1 + U2 + U3, data=data, subset=index, na.action = na.omit)
  b.nle_p_jail = lm(Y ~ X  + nle_p_jail*X + M + U1 + U2 + U3,  data=data, subset=index, na.action = na.omit)
  eff <- momed.eff.fn(a.nle_p_jail,b.nle_p_jail )
  effs <- c(effs, eff)
  aic <- c(nle_p_jail.aic = aic.fn(b.nle_p_jail))
  aics <- c(aics, aic)
  
  a.apq_p_cp =  lm(M ~ X + apq_p_cp*X + U1 + U2 + U3, data=data, subset=index, na.action = na.omit)
  b.apq_p_cp = lm(Y ~ X  + apq_p_cp*X + M + U1 + U2 + U3,  data=data, subset=index, na.action = na.omit)
  eff <- momed.eff.fn(a.apq_p_cp,b.apq_p_cp )
  effs <- c(effs, eff)
  aic <- c(apq_p_cp.aic = aic.fn(b.apq_p_cp))
  aics <- c(aics, aic)
  a.cpic_freq =  lm(M ~ X + cpic_freq*X + U1 + U2 + U3, data=data, subset=index, na.action = na.omit)
  b.cpic_freq = lm(Y ~ X  + cpic_freq*X + M + U1 + U2 + U3,  data=data, subset=index, na.action = na.omit)
  eff <- momed.eff.fn(a.cpic_freq,b.cpic_freq )
  effs <- c(effs, eff)
  aic <- c(cpic_freq.aic = aic.fn(b.cpic_freq))
  aics <- c(aics, aic)
  a.cpic_intensity =  lm(M ~ X + cpic_intensity*X + U1 + U2 + U3, data=data, subset=index, na.action = na.omit)
  b.cpic_intensity = lm(Y ~ X  + cpic_intensity*X + M + U1 + U2 + U3,  data=data, subset=index, na.action = na.omit)
  eff <- momed.eff.fn(a.cpic_intensity,b.cpic_intensity )
  effs <- c(effs, eff)
  aic <- c(cpic_intensity.aic = aic.fn(b.cpic_intensity))
  aics <- c(aics, aic)
  a.cpic_perc_threat =  lm(M ~ X + cpic_perc_threat*X + U1 + U2 + U3, data=data, subset=index, na.action = na.omit)
  b.cpic_perc_threat = lm(Y ~ X  + cpic_perc_threat*X + M + U1 + U2 + U3,  data=data, subset=index, na.action = na.omit)
  eff <- momed.eff.fn(a.cpic_perc_threat,b.cpic_perc_threat )
  effs <- c(effs, eff)
  aic <- c(cpic_perc_threat.aic = aic.fn(b.cpic_perc_threat))
  aics <- c(aics, aic)
  
  a.psi_pd_t =  lm(M ~ X + psi_pd_t*X + U1 + U2 + U3, data=data, subset=index, na.action = na.omit)
  b.psi_pd_t = lm(Y ~ X  + psi_pd_t*X + M + U1 + U2 + U3,  data=data, subset=index, na.action = na.omit)
  eff <- momed.eff.fn(a.psi_pd_t,b.psi_pd_t )
  effs <- c(effs, eff)
  aic <- c(psi_pd_t.aic = aic.fn(b.psi_pd_t))
  aics <- c(aics, aic)
  a.psi_total_t =  lm(M ~ X + psi_total_t*X + U1 + U2 + U3, data=data, subset=index, na.action = na.omit)
  b.psi_total_t = lm(Y ~ X  + psi_total_t*X + M + U1 + U2 + U3,  data=data, subset=index, na.action = na.omit)
  eff <- momed.eff.fn(a.psi_total_t,b.psi_total_t )
  effs <- c(effs, eff)
  aic <- c(psi_total_t.aic = aic.fn(b.psi_total_t))
  aics <- c(aics, aic)

  a.pattempts_any =  lm(M ~ X + pattempts_any*X + U1 + U2 + U3, data=data, subset=index, na.action = na.omit)
  b.pattempts_any = lm(Y ~ X  + pattempts_any*X + M + U1 + U2 + U3,  data=data, subset=index, na.action = na.omit)
  eff <- momed.eff.fn(a.pattempts_any,b.pattempts_any )
  effs <- c(effs, eff)
  aic <- c(pattempts_any.aic = aic.fn(b.pattempts_any))
  aics <- c(aics, aic)
  a.pattempts_thresh =  lm(M ~ X + pattempts_thresh*X + U1 + U2 + U3, data=data, subset=index, na.action = na.omit)
  b.pattempts_thresh = lm(Y ~ X  + pattempts_thresh*X + M + U1 + U2 + U3,  data=data, subset=index, na.action = na.omit)
  eff <- momed.eff.fn(a.pattempts_thresh,b.pattempts_thresh )
  effs <- c(effs, eff)
  aic <- c(pattempts_thresh.aic = aic.fn(b.pattempts_thresh))
  aics <- c(aics, aic)
  
  a.m_mood_disorder =  lm(M ~ X + m_mood_disorder*X + U1 + U2 + U3, data=data, subset=index, na.action = na.omit)
  b.m_mood_disorder = lm(Y ~ X  + m_mood_disorder*X + M + U1 + U2 + U3,  data=data, subset=index, na.action = na.omit)
  eff <- momed.eff.fn(a.m_mood_disorder,b.m_mood_disorder )
  effs <- c(effs, eff)
  aic <- c(m_mood_disorder.aic = aic.fn(b.m_mood_disorder))
  aics <- c(aics, aic)
  a.f_mood_disorder =  lm(M ~ X + f_mood_disorder*X + U1 + U2 + U3, data=data, subset=index, na.action = na.omit)
  b.f_mood_disorder = lm(Y ~ X  + f_mood_disorder*X + M + U1 + U2 + U3,  data=data, subset=index, na.action = na.omit)
  eff <- momed.eff.fn(a.f_mood_disorder,b.f_mood_disorder )
  effs <- c(effs, eff)
  aic <- c(f_mood_disorder.aic = aic.fn(b.f_mood_disorder))
  aics <- c(aics, aic)
  a.p_mood_disorder =  lm(M ~ X + p_mood_disorder*X + U1 + U2 + U3, data=data, subset=index, na.action = na.omit)
  b.p_mood_disorder = lm(Y ~ X  + p_mood_disorder*X + M + U1 + U2 + U3,  data=data, subset=index, na.action = na.omit)
  eff <- momed.eff.fn(a.p_mood_disorder,b.p_mood_disorder )
  effs <- c(effs, eff)
  aic <- c(p_mood_disorder.aic = aic.fn(b.p_mood_disorder))
  aics <- c(aics, aic)
  
  a.msubstance =  lm(M ~ X + msubstance*X + U1 + U2 + U3, data=data, subset=index, na.action = na.omit)
  b.msubstance = lm(Y ~ X  + msubstance*X + M + U1 + U2 + U3,  data=data, subset=index, na.action = na.omit)
  eff <- momed.eff.fn(a.msubstance,b.msubstance )
  effs <- c(effs, eff)
  aic <- c(msubstance.aic = aic.fn(b.msubstance))
  aics <- c(aics, aic)
  a.fsubstance =  lm(M ~ X + fsubstance*X + U1 + U2 + U3, data=data, subset=index, na.action = na.omit)
  b.fsubstance = lm(Y ~ X  + fsubstance*X + M + U1 + U2 + U3,  data=data, subset=index, na.action = na.omit)
  eff <- momed.eff.fn(a.fsubstance,b.fsubstance )
  effs <- c(effs, eff)
  aic <- c(fsubstance.aic = aic.fn(b.fsubstance))
  aics <- c(aics, aic)
  a.psubstance =  lm(M ~ X + psubstance*X + U1 + U2 + U3, data=data, subset=index, na.action = na.omit)
  b.psubstance = lm(Y ~ X  + psubstance*X + M + U1 + U2 + U3,  data=data, subset=index, na.action = na.omit)
  eff <- momed.eff.fn(a.psubstance,b.psubstance )
  effs <- c(effs, eff)
  aic <- c(psubstance.aic = aic.fn(b.psubstance))
  aics <- c(aics, aic)
  
  a.mpsychosis =  lm(M ~ X + mpsychosis*X + U1 + U2 + U3, data=data, subset=index, na.action = na.omit)
  b.mpsychosis = lm(Y ~ X  + mpsychosis*X + M + U1 + U2 + U3,  data=data, subset=index, na.action = na.omit)
  eff <- momed.eff.fn(a.mpsychosis,b.mpsychosis )
  effs <- c(effs, eff)
  aic <- c(mpsychosis.aic = aic.fn(b.mpsychosis))
  aics <- c(aics, aic)
  a.fpsychosis =  lm(M ~ X + fpsychosis*X + U1 + U2 + U3, data=data, subset=index, na.action = na.omit)
  b.fpsychosis = lm(Y ~ X  + fpsychosis*X + M + U1 + U2 + U3,  data=data, subset=index, na.action = na.omit)
  eff <- momed.eff.fn(a.fpsychosis,b.fpsychosis )
  effs <- c(effs, eff)
  aic <- c(fpsychosis.aic = aic.fn(b.fpsychosis))
  aics <- c(aics, aic)
  a.ppsychosis =  lm(M ~ X + ppsychosis*X + U1 + U2 + U3, data=data, subset=index, na.action = na.omit)
  b.ppsychosis = lm(Y ~ X  + ppsychosis*X + M + U1 + U2 + U3,  data=data, subset=index, na.action = na.omit)
  eff <- momed.eff.fn(a.ppsychosis,b.ppsychosis )
  effs <- c(effs, eff)
  aic <- c(ppsychosis.aic = aic.fn(b.ppsychosis))
  aics <- c(aics, aic)
  
  a.m_psychopathology =  lm(M ~ X + m_psychopathology*X + U1 + U2 + U3, data=data, subset=index, na.action = na.omit)
  b.m_psychopathology = lm(Y ~ X  + m_psychopathology*X + M + U1 + U2 + U3,  data=data, subset=index, na.action = na.omit)
  eff <- momed.eff.fn(a.m_psychopathology,b.m_psychopathology )
  effs <- c(effs, eff)
  aic <- c(m_psychopathology.aic = aic.fn(b.m_psychopathology))
  aics <- c(aics, aic)
  a.f_psychopathology =  lm(M ~ X + f_psychopathology*X + U1 + U2 + U3, data=data, subset=index, na.action = na.omit)
  b.f_psychopathology = lm(Y ~ X  + f_psychopathology*X + M + U1 + U2 + U3,  data=data, subset=index, na.action = na.omit)
  eff <- momed.eff.fn(a.f_psychopathology,b.f_psychopathology )
  effs <- c(effs, eff)
  aic <- c(f_psychopathology.aic = aic.fn(b.f_psychopathology))
  aics <- c(aics, aic)
  a.p_psychopathology =  lm(M ~ X + p_psychopathology*X + U1 + U2 + U3, data=data, subset=index, na.action = na.omit)
  b.p_psychopathology = lm(Y ~ X  + p_psychopathology*X + M + U1 + U2 + U3,  data=data, subset=index, na.action = na.omit)
  eff <- momed.eff.fn(a.p_psychopathology,b.p_psychopathology )
  effs <- c(effs, eff)
  aic <- c(p_psychopathology.aic = aic.fn(b.p_psychopathology))
  aics <- c(aics, aic)
  all <- c(med.model, effs, aics)
}


boot_obj <- boot(hbn.d, momed.fn, R=5000)

```

As we can see from our CI's, we did not detect any signifcant conditional indirect effects as hypothesized. But the absnece of a significant effect does not necessarily indicate the absence of results, as significance testing is only one way to assess the relevance of your model. We can also assess the model fits via AIC. It is worth noting, however, that the CIs on the paths that used the variables that we had the most biased sampling on, are the ones with the widest confidence intervals. This points to one of the important issues to keep in mind when employing the bootstrap: if you have not sourced your sample well, and the distribution on your variable is not well represented, then the bootstrap will shed light on the the lack of confidence you are able to have on your effects as a result.

### Conclusion

These model fits (although not exactly the appropriate statistic, as I am aware) are nevertheless revealing what I think is a pretty meaningful trend in this data. Our model fits, averaged across the 5000 bootstraps, are **considerably lower** and thus our model fit appears to be **much, much better**, with the parental stress items and the annual income variable, than with any of the parental absence or psychopathology variables. Interestingly as well, while the 3 CPIC (interparental conflict and discord) variables are all hanging out together in terms of general range of model fits, which reflects that this measure is likely doing a pretty good job of targeting it respective construct (and perhaps that maybe they hang a little too close together. And in fact a recent paper with this measure provided evidence for some higher order factor loadings that collapsed across these scales!), but we also take a also take a huge leap in in our model fits when examining these CPIC variables, comapared to the corporal punishment variables! We would expect these two items manifest similarly as both being threat variables! And more importantly, we would expect a better fit (reflect a more severe impact of the threat dimension of ELS exposur) with the more severe exposure, corporal punishment. Worth noting that corporal punishment is a highhly socially and culturally dependent construct, and we did not control for race in this analysis. 

As a small aside and commentary, what I actually this this observation is revealing is not necessarily that our  conceptual models are bad, but rather an issue in our interpretation and mathematical modeling them. The CPIC scales really long-term patterns of persistent conflict in the home. The Alabama Parenting Questionnaire corporal punishment items are less about the long term patterns of behavrios is more reflective of occasional parental behaviors (it's more of an inventory than assessment of severity of exposure). 

ELS researchers are either totally beholden to their theoretical models or completely divorced from it like with cumulative risk based models (those models only really only tell us one thing-- more stress is worse for developmental outcomes). There is however, a middle ground where a researcher is informed by models but still listens to the data. I would like to propose that threat and deprivation are inherently classifier variables while unpredictabiility is not. There are quantitative aspects of the early life everironment that neither threat not deprivation alone can fully capture. In the same respect, unpredictability does not necessarily reflect the emotional saliency or quality of the environmental experience, but rather the temporal and quantitatively distinguishable dynamics. I think are data often tries to tell us this, but we dont listen.  

Although our models could very well be bad, which, while terrifying, could be and important issue to probe. What would null hypothesis testing look like in studies probing that question? Perhaps, application of some nonparametric tests that intentionally probe the likelihood of our orginal hypotheses are neded. 

Finally, I'll reserve signifcance testing, if needed, for when I am able to more appropriately examine delta AIC. But this is , nevertheless, an entirely different way of approaching conceptual models of early life stress. And opening up the potential for lots of great future work!
