1. How does knowing data type (e.g., qualitative vs. quantitative) provide insights into the type of distribution a random variable comes from?

**Answer:** Random variables have the properties of the distributions they come from. That is nature of the variable is determined by the distribution it comes from. For example, knowing that a variable you are working with is qualitative and maybe, specifically, binary tells you, at minimum, that the distribution it comes from is definitely not normal. Contrastingly, if you are working with a continuous, quantitative variable it might be more likely to come from a normal distribution.

2. What exactly does it mean to say that a data point is a "random variable is normal with a mean of 100 and standard deviation of 15"?

**Answer:** The data point is a sample taken from a underlying normal distrbution in which the a most likely or expected value is 100 (the measure of central tendency) and values 15 above and 15 below 100 (between 85 and 115) have the highest probability of being observed.

4. Instead of rolling one die, run a set of experiments reporting the outcomes of rolling multiple dice at the same time.
        	a) Exp 1: 10,000 throws, 1 die
        	b) Exp 2: 10,000 throws, 2 dice
        	c) Exp 3: 10,000 throws, 3 dice
        	d) Exp 4; 10,000 throws, 6 dice.

Show the distribution of results for each experiment. Which of these experiments produces a distribution most similar to a normal normal distribution? Justify your conclusion using Q-Q plots.

**Answer:** Experiments 3 and 4 produce distributions most similar to the normal distribution. We can determine this by looking at the QQ plots for each experiment, which plots the our observed distribution in our data against a theoretical normal distribution. A straight line indicates that the observed distribution arises from a normal distribution. The QQ plots for exp 3 and 4 have a straight-ish line that curves at both ends. So these two experiements produce a probability distribution that is most similar to a normal distribution (but it's not great). 

```r
# -------------------------------
# Question 4:
# -------------------------------

#a) Exp 1: 10,000 throws, 1 die
b <- 6
dat <- data.frame(length=ceiling(runif(n, min=a, max=b)))
ggplot(dat, aes(length)) + geom_histogram(alpha = 0.9)
ggplot(dat, aes(length)) + geom_density(alpha = 0.2)
ggplot(dat, aes(sample=length)) + geom_point(stat = "qq") + ggtitle("Normal dist. vs. Exp 1: 10,000 throws, 1 die")

#b) Exp 2: 10,000 throws, 2 dice
b <- 6^2
dat <- data.frame(length=ceiling(runif(n, min=a, max=b)))
ggplot(dat, aes(length)) + geom_histogram(alpha = 0.9)
ggplot(dat, aes(length)) + geom_density(alpha = 0.2)
ggplot(dat, aes(sample=length)) + geom_point(stat = "qq") + ggtitle("Normal dist. vs. Exp 2: 10,000 throws, 2 die")

#c) Exp 3: 10,000 throws, 3 dice
b <- 6^3
dat <- data.frame(length=ceiling(runif(n, min=a, max=b)))
ggplot(dat, aes(length)) + geom_histogram(alpha = 0.9)
ggplot(dat, aes(length)) + geom_density(alpha = 0.2)
ggplot(dat, aes(sample=length)) + geom_point(stat = "qq") + ggtitle("Normal dist. vs. Exp 3: 10,000 throws, 3 die")

#d) Exp 4; 10,000 throws, 6 dice.
b <- 6^6
dat <- data.frame(length=ceiling(runif(n, min=a, max=b)))
ggplot(dat, aes(length)) + geom_histogram(alpha = 0.9)
ggplot(dat, aes(length)) + geom_density(alpha = 0.2)
ggplot(dat, aes(sample=length)) + geom_point(stat = "qq") + ggtitle("Normal dist. vs. Exp 4: 10,000 throws, 6 die")
```
